## <u>Mod√®le th√©orique</u>

Nous √©tudions un syst√®me de spins quantiques $S = \frac{1}{2}$ sur un r√©seau, soumis √† un champ magn√©tique externe $h$ dirig√© selon l‚Äôaxe $z$. L‚ÄôHamiltonien g√©n√©ral du mod√®le s‚Äô√©crit :

$$
H = \sum_{\langle i,j \rangle} J_{ij} \vec{S}_i \cdot \vec{S}_j - h \sum_i S_i^z
$$

o√π :
- $\vec{S}_i$ est l‚Äôop√©rateur de spin sur le site $i$,
- $J_{ij}$ est le couplage entre les spins situ√©s aux sites $i$ et $j$,
- $h$ est l‚Äôintensit√© du champ magn√©tique appliqu√©,
- la somme $\langle i,j \rangle$ porte sur les paires de sites voisins.

Ce mod√®le capture l‚Äôessentiel de la comp√©tition entre les effets d‚Äô√©change (tendance √† l‚Äôordre magn√©tique) et les effets du champ externe (tendance √† l‚Äôalignement des spins).

---

### Grandeurs thermodynamiques d‚Äôint√©r√™t

Nous nous int√©ressons principalement √† deux observables macroscopiques d√©riv√©es de la fonction de partition :

- **La chaleur sp√©cifique** (√† champ $h$ fix√©) :

$$
C_v(T, h) = \left( \frac{\partial E}{\partial T} \right)_h = \frac{\partial}{\partial T} \langle H \rangle
$$

Elle mesure les fluctuations d‚Äô√©nergie du syst√®me et donne des informations sur les transitions thermiques ou les excitations internes.

- **La susceptibilit√© magn√©tique** (√† temp√©rature $T$ fix√©e) :

$$
\chi(T, h) = \left( \frac{\partial m}{\partial h} \right)_T
$$

avec la magn√©tisation d√©finie comme :

$$
\vec{m} = \sum_i \langle \vec{S}_i \rangle
$$

La susceptibilit√© quantifie la r√©ponse lin√©aire du syst√®me √† un faible champ magn√©tique et est particuli√®rement sensible aux corr√©lations magn√©tiques √† basse temp√©rature.

---

Dans la suite, ces quantit√©s seront calcul√©es √† partir des √©tats propres de l‚ÄôHamiltonien restreint √† des clusters finis, puis moyenn√©es sur la distribution statistique induite par la dilution du r√©seau.


## <u> Cas d'un cluster compos√© d‚Äôun seul site occup√© </u> 

Consid√©rons un cluster constitu√© d‚Äôun unique site de spin $S = \frac{1}{2}$, coupl√© √† deux voisins absents (vides). Dans ce cas, l‚Äôinteraction d‚Äô√©change est nulle car il n‚Äôy a aucun site voisin avec lequel interagir :

$$
H = -h S_{p_z} \quad \text{puisque} \quad \sum_{\langle i,j \rangle} J_{ij} S_i S_j = 0
$$

L‚Äô√©nergie moyenne est d√©finie par :

$$
E = \langle H \rangle = -h \langle S_{p_z} \rangle
$$

On rappelle les expressions usuelles de la susceptibilit√© magn√©tique et de la chaleur sp√©cifique :

$$
\chi(T,h) = \frac{\partial}{\partial h} \langle S_{p_z} \rangle
$$

$$
C_v = \frac{\partial \langle E \rangle}{\partial T}
$$

## Calcul explicite pour un spin $1/2$ 

Les √©tats propres du spin $S_{p_z}$ sont $\pm \frac{1}{2}$, avec les √©nergies correspondantes :

$$
E_{\pm} = \mp \frac{h}{2}
$$

La fonction de partition est :

$$
Z = e^{-\beta E_{-}} + e^{-\beta E_{+}} = e^{\beta \frac{h}{2}} + e^{-\beta \frac{h}{2}} = 2 \cosh\left( \frac{\beta h}{2} \right)
$$

L‚Äô√©nergie moyenne est obtenue par d√©rivation de $\ln Z$ :

$$
\langle E \rangle = -\frac{\partial \ln Z}{\partial \beta} = -\frac{h}{2} \tanh\left( \frac{\beta h}{2} \right)
$$

D‚Äôo√π la magn√©tisation :

$$
\langle S_{p_z} \rangle = \frac{1}{2} \tanh\left( \frac{\beta h}{2} \right)
$$

###  Susceptibilit√© magn√©tique 

En d√©rivant $\langle S_{p_z} \rangle$ par rapport √† $h$ :

$$
\chi(T, h) = \frac{\partial \langle S_{p_z} \rangle}{\partial h} = \frac{\beta}{4} \left[1 - \tanh^2\left( \frac{\beta h}{2} \right) \right]
$$

Dans la limite $h \to 0$ :

$$
\chi(T) = \frac{1}{4k_B T}
$$

### Chaleur sp√©cifique 

En d√©rivant $\langle E \rangle$ par rapport √† $T$ :

$$
C_v = \frac{\partial \langle E \rangle}{\partial T} = -\frac{\partial \langle E \rangle}{\partial \beta} \cdot \frac{\partial \beta}{\partial T} = \beta^2 \frac{\partial \langle E \rangle}{\partial \beta}
$$

Or :

$$
\frac{\partial \langle E \rangle}{\partial \beta} = \frac{h^2}{4} \left[1 - \tanh^2\left( \frac{\beta h}{2} \right) \right]
$$

Donc :

$$
C_v = \frac{\beta^2 h^2}{4} \left[1 - \tanh^2\left( \frac{\beta h}{2} \right) \right]
$$

Dans la limite $h \to 1$, on obtient la chaleur sp√©cifique d‚Äôun spin $1/2$ en champ unitaire.


## <u> Cas d'un cluster compos√© de 2 sites occup√©s </u> 

Consid√©rons un cluster compos√© de deux spins $S = \frac{1}{2}$ coupl√©s entre eux, et soumis √† un champ magn√©tique $h$. L‚ÄôHamiltonien s‚Äô√©crit :

$$
H = -J \, \vec{S}_{p_1} \cdot \vec{S}_{p_2} - h(S_{p_1,z} + S_{p_2,z})
$$

Ce Hamiltonien comprend :
- une interaction d‚Äô√©change entre les deux spins avec une constante $J$ ;
- un couplage Zeeman avec un champ magn√©tique externe $h$.

## Diagonalisation dans la base des √©tats propres du spin total

On pose :

$$
\vec{S}_{\text{tot}} = \vec{S}_{p_1} + \vec{S}_{p_2} \quad \Rightarrow \quad \vec{S}_{p_1} \cdot \vec{S}_{p_2} = \frac{1}{2} \left( S_{\text{tot}}^2 - S_{p_1}^2 - S_{p_2}^2 \right)
$$

Puisque $S_{p_1}^2 = S_{p_2}^2 = \frac{3}{4}$ pour des spins $1/2$, on a :

$$
\vec{S}_{p_1} \cdot \vec{S}_{p_2} = \frac{1}{2} \left( S_{\text{tot}}(S_{\text{tot}} + 1) - \frac{3}{2} \right)
$$

Les √©tats propres du syst√®me sont les singulets ($S=0$) et triplets ($S=1$) :

- Triplet :
  $$
  \begin{aligned}
  &\left|1,1\right\rangle = \left|\uparrow \uparrow \right\rangle \\
  &\left|1,0\right\rangle = \frac{1}{\sqrt{2}}\left( \left|\uparrow \downarrow \right\rangle + \left|\downarrow \uparrow \right\rangle \right) \\
  &\left|1,-1\right\rangle = \left|\downarrow \downarrow \right\rangle
  \end{aligned}
  $$

- Singulet :
  $$
  \left|0,0\right\rangle = \frac{1}{\sqrt{2}}\left( \left|\uparrow \downarrow \right\rangle - \left|\downarrow \uparrow \right\rangle \right)
  $$

## √ânergies propres

Utilisons $S_{\text{tot}}^2$ pour √©valuer l‚Äô√©nergie de chaque √©tat :

- **Triplets** $(S_{\text{tot}} = 1)$ :
  $$
  \vec{S}_{p_1} \cdot \vec{S}_{p_2} = \frac{1}{2}(1(1+1) - \frac{3}{2}) = \frac{1}{4}
  $$

  Les √©nergies sont donc :
  $$
  \begin{aligned}
  &E_{1,1} = -J \cdot \frac{1}{4} - h \cdot 1 = -\frac{J}{4} - h \\
  &E_{1,0} = -\frac{J}{4} \\
  &E_{1,-1} = -\frac{J}{4} + h
  \end{aligned}
  $$

- **Singulet** $(S_{\text{tot}} = 0)$ :
  $$
  \vec{S}_{p_1} \cdot \vec{S}_{p_2} = \frac{1}{2}(0(0+1) - \frac{3}{2}) = -\frac{3}{4}
  $$

  Donc :
  $$
  E_{0,0} = +\frac{3J}{4}
  $$

## Fonction de partition

$$
Z = e^{-\beta E_{1,1}} + e^{-\beta E_{1,0}} + e^{-\beta E_{1,-1}} + e^{-\beta E_{0,0}} = e^{\beta(\frac{J}{4} + h)} + e^{\beta \frac{J}{4}} + e^{\beta(\frac{J}{4} - h)} + e^{-\beta \frac{3J}{4}}
$$

## Observables thermodynamiques

### Magn√©tisation

On d√©finit la magn√©tisation totale :

$$
\langle S_{\text{tot},z} \rangle = \frac{1}{Z} \left[ (+1) e^{\beta(\frac{J}{4} + h)} + (0) e^{\beta \frac{J}{4}} + (-1) e^{\beta(\frac{J}{4} - h)} + 0 \cdot e^{-\beta \frac{3J}{4}} \right] = \frac{e^{\beta(\frac{J}{4} + h)} - e^{\beta(\frac{J}{4} - h)}}{Z}
$$

### Susceptibilit√©

Par d√©finition :

$$
\chi = \frac{\partial \langle S_{\text{tot},z} \rangle}{\partial h}
$$

Le calcul explicite donne une expression de la forme :

$$
\chi = \beta \left[ \langle S_{\text{tot},z}^2 \rangle - \langle S_{\text{tot},z} \rangle^2 \right]
$$

### Chaleur sp√©cifique

√Ä partir de l‚Äô√©nergie moyenne :

$$
\langle E \rangle = \frac{1}{Z} \sum_i E_i e^{-\beta E_i}
$$

et de sa d√©riv√©e :

$$
C_v = \frac{\partial \langle E \rangle}{\partial T} = \beta^2 \left( \langle E^2 \rangle - \langle E \rangle^2 \right)
$$

Cette derni√®re formule est celle d√©riv√©e du th√©or√®me de fluctuation de l‚Äô√©nergie.

---

**Remarque :** on retrouve les m√™mes structures que dans le cas d‚Äôun seul spin, mais avec plusieurs niveaux d‚Äô√©nergie et un enrichissement des comportements thermodynamiques.




On commence par definir deux fonction `Cv1` et `chi1` qui vont nous permettre d'afficher les courbes de $C_v$ et $\chi$ grace au fonction que nous avons calculer pr√©c√©dement.

```python
# Constantes
kB = 1
h = 0
T = np.linspace(0.1, 3.5, 500)
beta = 1 / (kB * T)

# === R√©seau √† 1 particule ===
def chi_1(T):
    return (1 - np.tanh(h/(kB*T))**2) / (4 * kB * T)

def Cv_1(T):
    beta_1 = 1 / (kB * T)
    return (beta_1**2 * h**2 / 4) * (1 - np.tanh(beta_1 * h / 2)**2)
```

Nous allons aussi cr√©e deux fonctions pour le cas du cluster a 2 particules, comme pr√©c√©dement en utilisant les fonctions calcul√© theoriquement. Nous cr√©ons aussi une fonction qui va nous renvoyer les valeurs d'energies pour diff√©rents $J$ afin d'√©tudier le cas ferromagnetique et antiferromagnetique.

```python
# === Fonction g√©n√©rique pour 2 particules en interaction ===
def calc_spin2(J, beta):
    E = np.array([
        J/4 - h,   # triplet Sz = +1
        J/4,       # triplet Sz = 0
        J/4 + h,   # triplet Sz = -1
       -3*J/4      # singulet
    ])
    Sz = np.array([1, 0, -1, 0])
    
    Z = np.sum(np.exp(-beta[:, None] * E), axis=1)
    E_moy = np.sum(E * np.exp(-beta[:, None] * E), axis=1) / Z
    E2_moy = np.sum(E**2 * np.exp(-beta[:, None] * E), axis=1) / Z
    Cv = kB * beta**2 * (E2_moy - E_moy**2)
    
    Sz_moy = np.sum(Sz * np.exp(-beta[:, None] * E), axis=1) / Z
    Sz2_moy = np.sum(Sz**2 * np.exp(-beta[:, None] * E), axis=1) / Z
    chi = beta * (Sz2_moy - Sz_moy**2)
    
    return Cv, chi
```

On affiche ensuite les fonctions.

```python
# Calculs pour J = -1 (ferromagn√©tique) et J = +1 (antiferromagn√©tique)
Cv_Jp, chi_Jp = calc_spin2(J=1.0, beta=beta)
Cv_Jm, chi_Jm = calc_spin2(J=-1.0, beta=beta)

# === Trac√©s ===

# Susceptibilit√©
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(T, chi_1(T), '--', label='œá (1 particule)', color='gray')
plt.plot(T, chi_Jp, label='œá (2 particules J = +1)', color='blue')
plt.plot(T, chi_Jm, label='œá (2 particules J = -1)', color='green')
plt.xlabel('T')
plt.ylabel('œá')
plt.title('Susceptibilit√© magn√©tique')
plt.legend()
plt.grid(True)

# Capacit√© calorifique
plt.subplot(1, 2, 2)
plt.plot(T, Cv_1(T), '--', label='Cv (1 particule)', color='gray')
plt.plot(T, Cv_Jp, label='Cv (2 particules J = +1)', color='red')
plt.plot(T, Cv_Jm, label='Cv (2 particules J = -1)', color='orange')
plt.xlabel('T')
plt.ylabel('Cv')
plt.title('Capacit√© calorifique')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Cha√Æne 1D infinie avec dilution al√©atoire

Nous consid√©rons un syst√®me infini de spins $S = \frac{1}{2}$ sur une cha√Æne unidimensionnelle, o√π chaque site est occup√© avec une probabilit√© $p$. Cette dilution al√©atoire donne naissance √† une distribution de **clusters de spins** s√©par√©s par des sites vides.

---

## Types de clusters et poids associ√©s

- **Site vide seul** :
  - Poids : $q = 1 - p$
  - Partition : $Z_0 = 1$ (aucune contribution √©nerg√©tique)

- **Cluster d‚Äôun spin isol√© entre deux vides** :
  - Poids : $p(1 - p)^2$
  - Partition : $Z_1 = 2 \cosh\left( \frac{\beta h}{2} \right)$

- **Cluster de deux spins coupl√©s, encadr√©s par des sites vides** :
  - Poids : $p^2 (1 - p)^2$
  - Partition : $Z_2 = e^{\beta \frac{3J}{4}} + 2 e^{-\beta \frac{J}{4}} \cosh(\beta h)$

- **Clusters plus grands** (taille $n$) :
  - Poids : $P(n) = n (1 - p)^2 p^n$
  - Partition : $Z_n$ (calcul√©e num√©riquement √† partir du Hamiltonien)

---

## Partition moyenne par site

On d√©finit une fonction de partition moyenne (logarithmique) par site :

$$
\frac{\ln \overline{Z}}{N} = \sum_n P(n) \ln Z_n
$$

avec :

$$
P(n) = n (1 - p)^2 p^n
$$

---

## Quantit√©s thermodynamiques par site

- **√ânergie moyenne** :

$$
\left\langle \frac{E}{N} \right\rangle = -\frac{\partial}{\partial \beta} \left( \frac{\ln \overline{Z}}{N} \right) = \sum_n P(n) \left\langle E_n \right\rangle
$$

- **Chaleur sp√©cifique** :

$$
\frac{C_v}{N} = \sum_n P(n) \cdot C_{v,n}
$$

- **Susceptibilit√© magn√©tique** :

$$
\frac{\chi}{N} = \sum_n P(n) \cdot \chi_n
$$

---

## Impl√©mentation num√©rique

Nous calculons num√©riquement $Z_n$, $E_n$, $C_{v,n}$, $\chi_n$ pour chaque cluster de taille $n$ jusqu‚Äô√† un ordre donn√© (ex : $n = 8$), puis reconstruisons les quantit√©s extensives moyennes via les pond√©rations $P(n)$.

---

Nous utilisons les m√™mes fonctions que pr√©c√©dement pour le calcul de $C_v$ et $\chi$ pour les cas de cluster de taille 1 et 2, et nous allons donc calculer ces fonctions en les pond√©rants par leurs poids. 

Nous commencons par d√©finir les param√®tres de notre simulation :

```python
# Temp√©rature et beta
T_vals = np.linspace(0.1, 10, 300)
beta_vals = 1 / (kB * T_vals)

# Param√®tres
p_vals = [0.05, 0.1, 0.2]
J_vals = [-1, 0, 1]
```

Ensuite nous affichons les r√©sultats pour diff√©rents $p$ et $J$

```python
def calc_spin2(J, beta):
    E = np.array([
        J/4 - h,
        J/4,
        J/4 + h,
        -3*J/4
    ])
    Sz = np.array([1, 0, -1, 0])
    
    Z = np.sum(np.exp(-beta[:, None] * E), axis=1)
    E_moy = np.sum(E * np.exp(-beta[:, None] * E), axis=1) / Z
    E2_moy = np.sum(E**2 * np.exp(-beta[:, None] * E), axis=1) / Z
    Cv = kB * beta**2 * (E2_moy - E_moy**2)
    
    Sz_moy = np.sum(Sz * np.exp(-beta[:, None] * E), axis=1) / Z
    Sz2_moy = np.sum(Sz**2 * np.exp(-beta[:, None] * E), axis=1) / Z
    chi = beta * (Sz2_moy - Sz_moy**2)
    
    return Cv, chi

# Cr√©ation du plot
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(12, 10), sharex=True)

for i, J in enumerate(J_vals):
    for p in p_vals:
        Cv_2, chi_2 = calc_spin2(J, beta_vals)
        Cv_total = (p * (1 - p)**2 * Cv_1(T_vals) + 2 * p**2 * (1 - p)**2 * Cv_2) / 2
        chi_total = (p * (1 - p)**2 * chi_1(T_vals) + 2 * p**2 * (1 - p)**2 * chi_2) / 2
        
        axs[i, 0].plot(T_vals, Cv_total, label=f"p={p}")
        axs[i, 1].plot(T_vals, chi_total, label=f"p={p}")
    
    axs[i, 0].set_ylabel('$C_v/N$')
    axs[i, 1].set_ylabel('$œá/N$')
    axs[i, 0].set_title(f'J = {J}', loc='left', fontsize=10)
    axs[i, 0].grid(True)
    axs[i, 1].grid(True)
    if i == len(J_vals) - 1:
        axs[i, 0].set_xlabel('Temp√©rature $T$')
        axs[i, 1].set_xlabel('Temp√©rature $T$')

axs[0, 0].legend()
axs[0, 1].legend()

fig.suptitle("√âvolution de $C_v$ et $œá$ en fonction de T pour un d√©veloppement √† l'ordre 2", fontsize=14)
plt.tight_layout()
plt.show()
```

- $\frac{\chi}{N}$ en fonction de $T$

Ces r√©sultats sont en accord avec les expressions analytiques dans les cas limites (faible $p$ ou $p \to 1$).
