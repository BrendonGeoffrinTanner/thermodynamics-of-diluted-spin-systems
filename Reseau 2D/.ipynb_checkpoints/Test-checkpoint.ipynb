# Generation des clusters 

Nous commencons par vouloir generer tout les cluster connexes de tailles $N$, pour cela nous allons devoir en premier definir un type de voisinages sur nos reseau, et utiliser une méthode de generations de nos cluster sur ce type de reseau, nous allons ici seulement nous interessée au cas du reseau carrée et triangulaire.

Commencons par définir ce que nous allons appeller un cluster, un cluster est un ensemble de points du reseau ou chaque points est connecté à au moins un autre point par une liaison du reseau.

Nous devons donc crée 2 programmes qui vont devoir generer pour les différents type de reseau :
* Le nombres de site dans le cluster
* Les liaisons entre chaque site
* Le nombre de voisin vide, nommée ici voisin exterieur au cluster
* Le nombre de configuration de notre cluster, le nombre de facon différentes que l'on a de representer notre cluster

## Type de voisinages 

Dans le cas d'un reseau carré chaque point est connectées seulement au voisin avec qui il partage une arrête, on va donc pour ce type de reseau utiliser un reseau de Von Neumann, en utilisant les coordonnées $(x,y)$

![Voisinage de Von Neumann](https://www.researchgate.net/publication/347771430/figure/fig4/AS:1004055048503300@1616396786798/on-Neumann-and-Moore-neighborhoods.png)

Nous allons dans le cas du reseau triangle utiliser aussi un reseau de Von Neumann mais adapté au reseau triangulaire. On utilisera donc les coordonnées $(x,y)$ transformée pour ce type de reseau.

## Depth First Search

Depth-First Search _(DFS)_ ou parcours en profondeur est un algortihme recursif permettant d'explorer un espace de manière exhaustive. A partir d'un point initial on ava,ce le plus loin possivle dans une direction avant de revenir en arrière pour explorer d'autres possibilités.

<u>Explication des étapes du programme :</u>
- Initialisation : Le parcours débute à partir d'un point dentral du reseau
- Exploration récursive : On ajoute un voivin valide au cluster actuel et on poursuit récursivement l'exploration
- Condition d'arret : L'orsqu'un cluster contient exactement $N$ points il est convertie en une forme canonique invariante par rotation, symetrie, translation et est ajouté si il est unique

## Forme cannonique d'un cluster

Avec une transformation canonique on représente chaque custer sous une forme unique indépendante de sa position afin d'éviter les doublons.

On commence par faire une translation du cluster vers l'origine. Soit $C = \{(x_i,y_i)\}^N_{i=1}$ un cluster de $N$ points sur une grille. On commence par translater tous les points afin que le point en haut a gauche du cluster soit deplacer en $(0,0)$.

$$
x_{min} = \min_{(x,y) \in C}(x) \quad y_{min} = \min_{(x,y) \in C}(y)
$$

Notre cluster a de nouvelle coordonnées :

$$
C' = \{(x-x_{min},y-y_{min}) \ | \ (x,y) \in C\}
$$

Sur un reseau carrée un cluster peut être transformé par :
* Une rotation de $0° = 360°$,$90°$,$180°$ ou $270°$
* Une symétrie honrizontale ou vertivale selon les axes du reseau

Sur un reseau triangle un cluster peut être transformé par :
* Une rotation de $0° = 360°$,$60°$,$120°$,$180°$,$240°$ ou $300°$
* Une symétrie honrizontale ou vertivale selon les axes du reseau


Pour chaque transformation geometrique $T_k$ on applique : 
$$ C_k = \text{normalize}(T_k(C'))$$

Ou normalize signifie à nouveau ramener le cluster transformé en haut a gauche.

Apres avoir appliqué toutes les transformations géometrique valides on obtient un ensemble de version équivalentes du cluster. Chaque version transformées est un ensemble de couples de coordonnées entières :

$$
C_k = \{(x_i^{(k)},y_i^{(k)})\}^N_{i=1}
$$

Pour chaque $k$, les point sont triées selon l'ordre l'exicographie, c'est a dire que l'on compare la première coordonnées, ici $x$ et i les $x$ sont éaux on compare ensuite la seconde coordonnées.

$$
(x_1^{(k)},y_1^{(k)}) < (x_2^{(k)},y_2^{(k)}) < ...< (x_N^{(k)},y_N^{(k)})
$$

L'ensemble des représentations transformées et alors : $T = \{ sort(C_k) \ | \ \text{$T_k$ est une rot/sym de }C  \}$. La forme canonique est donc définie comme la plus petite au sens lexicographique représentation parmit toute celle générée. Les cluster sont donc générée et afficher via les programme `Carree/graphique_cluster_reseau_carre.py` pour le reseau carrée et `Triangle/graphique_cluster_reseau_triangle.py` pour le reseau triangle. L'affiche de chaque cluster ainsi que leurs nombre de voisin et leurs nombre de voisin son utile car il nous permet de vérifier les résultats a la main pour des petits $N$.
